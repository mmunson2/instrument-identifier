```python
# Extracting data for each field
epochs = [data["epoch"] for data in performance]
accuracy = [data["Accuracy"] for data in performance]
threshold_accuracy = [data["Threshold_Accuracy"] for data in performance]
# rmse = [data["RMSE"] for data in performance]

# Plotting each field
plt.plot(epochs, accuracy, label="Accuracy", marker='o', color='green')
plt.plot(epochs, threshold_accuracy, label="Threshold Accuracy", marker='o', color='orange')
# plt.plot(epochs, rmse, label="RMSE")

# Adding labels and legend
plt.xlabel("Epoch Count")
plt.ylabel("Percent Accuracy")
plt.title("Daytime Model Performance over Epochs")
plt.legend()
plt.grid(True)

plt.xticks(epochs)

plt.savefig('VS_Accuracy_Over_Epoch_Daytime.png', dpi=300, bbox_inches='tight')

# Show plot
plt.show()
```
